```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('earthquake.csv')
    

normal_df = df[df['tsunami'] == 0]
anomalous_df = df[df['tsunami'] == 1]
```

```python
features = ['magnitude', 'cdi', 'mmi', 'dmin', 'gap', 'depth']

plt.figure(figsize=(15, 10))
# Plot histograms for each feature
    
for i, feature in enumerate(features, 1):
    plt.subplot(2, 4, i)
    sns.histplot(normal_df[feature], kde=True, color='blue', label='Normal', stat='density',bins=30)
    sns.histplot(anomalous_df[feature], kde=True, color='red', label='Anomalous',stat='density', bins=30)
    plt.title(f'Distribution of {feature}')
    plt.legend()
    
plt.tight_layout()
plt.show()
```
